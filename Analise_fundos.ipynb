```python
import investpy as inv
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['figure.figsize'] = (16,8)
```

```python
fundos_br = inv.get_funds_list(country='Brazil')
```

```python
pesq = inv.search_funds(by='name', value='icatu van')
pesq
```

```python
pesq['name'][10]
```

```python
fundo = str(pesq['name'][10])
```

```python
icatu = inv.get_fund_historical_data(fundo, country='brazil', from_date='01/01/1900', to_date='10/12/2020')['Close']
icatu
```

```python
sns.set()
icatu.plot(title=pesq['name'][10])
plt.legend()
plt.axes

plt.savefig('Icatu.jpg', format='png', bbox_inches='tight')
```
